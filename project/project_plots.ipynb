## Balanced network recieving correlated Poisson input -- change $V_T$

```python
# Load the .mat file. Load the recorded variables from the matlab simulation.

mat_contents=sio.loadmat('./f_i_curve_network_VT.mat')

# Save some variables from mat file.
mean_e_rate = mat_contents['mean_e_rate']
mean_i_rate = mat_contents['mean_i_rate']
rx_vector = mat_contents['rx_vector'].squeeze()
threshold_vector = mat_contents['threshold_vector'].squeeze()
num_points_input = mat_contents['num_points_input'].squeeze()
num_points_gain = mat_contents['num_points_gain'].squeeze()
mRee = mat_contents['mRee']
mRei = mat_contents['mRei']
mRii = mat_contents['mRii']
mean_corrs = (mRee + 2*mRei + mRii)/4
mean_rate = 0.8*mean_e_rate + 0.2*mean_i_rate
```

```python
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.4, rc={"lines.linewidth":3.3})
    
plt.plot(1000*rx_vector,1000*mean_rate[:,0],label='$V_T=$%.1d mV'%threshold_vector[0])
plt.plot(1000*rx_vector,1000*mean_rate[:,1],label='$V_T=$%.1d mV'%threshold_vector[1])
plt.plot(1000*rx_vector,1000*mean_rate[:,2],label='$V_T=$%.1d mV'%threshold_vector[2])


plt.xlabel('Input rate (Hz)')
plt.ylabel('Output rate (Hz)')
plt.legend()

sns.despine()
plt.show()
```

```python
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.4, rc={"lines.linewidth":3.3})
    
plt.plot(1000*rx_vector,mean_corrs[:,0],label='$V_T=$%.1d mV'%threshold_vector[0])
plt.plot(1000*rx_vector,mean_corrs[:,1],label='$V_T=$%.1d mV'%threshold_vector[1])
plt.plot(1000*rx_vector,mean_corrs[:,2],label='$V_T=$%.1d mV'%threshold_vector[2])


plt.ylabel('Correlations')
plt.xlabel('Input rate (Hz)')
plt.legend()

sns.despine()
plt.show()
```

```python
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.4, rc={"lines.linewidth":3.3})
    
plt.plot(threshold_vector,1000*mean_rate[0,:],color='tab:brown')


plt.xlabel('$V_T$ (mV)')
plt.ylabel('Output rate (Hz)')
# plt.legend()

sns.despine()
plt.show()
```

```python
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.4, rc={"lines.linewidth":3.3})
    
plt.plot(threshold_vector,mean_corrs[0,:],color='tab:pink')


plt.xlabel('$V_T$ (mV)')
plt.ylabel('Mean correlations')
# plt.legend()

sns.despine()
plt.show()
```
