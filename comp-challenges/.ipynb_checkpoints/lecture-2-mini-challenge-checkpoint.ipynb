```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
V1 = [0] * (10**5)
V2 = [0] * (10**5)
V3 = [0] * (10**5)

V0=3

for i in range(len(V1)):
    V1[i] = (65539*V0)%(2**31)
    V2[i] = (65539*V1[i])%(2**31)
    V3[i] = (65539*V2[i])%(2**31)

    V0=V3[i]

ax = plt.axes(projection='3d')

ax.scatter3D(V1,V2,V3,'black')
ax.view_init(60,65)
```
