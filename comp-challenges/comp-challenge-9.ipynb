In this challenge you will implement a spatial version of the rock–paper–scissors game. The
agents are placed on a lattice with periodic boundary conditions, and each agent uses one of
three strategies. We can thus think of the agents as belonging to one of three species. Each
cell (location) in the lattice is either empty or occupied by an agent of one particular species.
Start with populating the lattice with agents of each type, and leaving some cells empty. The
simulation proceeds by picking a cell uniformly at random in the lattice, and then picking
a random neighboring cell. If both cells are empty, or are occupied by agents of the same
species, nothing happens. Otherwise three things can happen:

1. If only one of the cells is occupied, the agent in the occupied cell will reproduce. The descendant belongs to the same species, ie uses the same strategy.

2. If the two cells are occupied by different species, then with probability $p$ they fight. The agent with the losing strategy dies, and their cell is vacated.

3. With probability $1−p$, the agents in the two cells swap places. For this exercise it is best to animate your output. You can use FuncAnimation from matplotlib, but there are other choices. Start with a 30 by 30 grid at least.

(a) Illustrate what happens for high values of $p$, and then decrease $p$. You should see the
formation of spiral structures that increase in size as mobility is increased.

(b) Compute the time series for the frequency for each species, and plot them in time. How
does this change with $p$?

(c) Assume that there is a higher rate of reactions in the upper left quadrant of the lattice,
so that the probability of choosing a cell in this corner has probability is twice as likely
as picking a cell from another quadrant. How do the results in part (a) change?

## Define the function for simulating the agent based model

```python
class agent_based_model:
    """
    Agent based model fo a lattice with unoccupied spaces and spaces occupied by rock, paper, or scissors
    """
    def __init__(self, space_lattice,N,p_fight, seed=None):

        if seed is not None:
            np.random.seed(seed)

        self.space_lattice = space_lattice
        self.N = N
        self.p_fight = p_fight
        
    def step(self,space_lattice):
        """
        Select a or A for birth.
        Select a or A for death.

        Update history and states.
        """
        randomRow = np.random.randint(self.N, size=1)
        randomColumn = np.random.randint(self.N, size=1)
        
        
        move_hor = np.random.randint(-1,1)
        move_ver = np.random.randint(-1,1)
        
        while move_hor == 0 and move_ver ==0:
            move_hor = np.random.randint(-1,1)
            move_ver = np.random.randint(-1,1)
        
        neighborRow = randomRow + move_hor
        neighborColumn = randomColumn + move_ver
        
        if neighborRow == -1:
            neighborRow = N-1
        if neighborRow == N:
            neighborRow = 0
        if neighborColumn == -1:
            neighborColumn = N-1
        if neighborColumn == N:
            neighborColumn = 0
            
#         print(randomRow, randomColumn)
#         print(neighborRow, neighborColumn)
        
        
        U = np.random.uniform(0,1)
        
        # If we pick an occupied cell, and the chosen neighboring cell is unoccupied, reproduce the same agent.
        if self.space_lattice[randomRow,randomColumn]>0 and self.space_lattice[neighborRow,neighborColumn]==0:
            self.space_lattice[neighborRow,neighborColumn] = self.space_lattice[randomRow,randomColumn]
        if self.space_lattice[randomRow,randomColumn]>0 and self.space_lattice[neighborRow,neighborColumn]>0:
            if p_fight >= U: # If the two cells fight
                if self.space_lattice[randomRow,randomColumn]==1 and self.space_lattice[neighborRow,neighborColumn]==2:
                    self.space_lattice[randomRow,randomColumn] = 0
                elif self.space_lattice[randomRow,randomColumn]==2 and self.space_lattice[neighborRow,neighborColumn]==1:
                    self.space_lattice[neighborRow,neighborColumn] = 0
                elif self.space_lattice[randomRow,randomColumn]==1 and self.space_lattice[neighborRow,neighborColumn]==3:
                    self.space_lattice[neighborRow,neighborColumn] = 0
                elif self.space_lattice[randomRow,randomColumn]==3 and self.space_lattice[neighborRow,neighborColumn]==1:
                    self.space_lattice[randomRow,randomColumn] = 0
                elif self.space_lattice[randomRow,randomColumn]==2 and self.space_lattice[neighborRow,neighborColumn]==3:
                    self.space_lattice[randomRow,randomColumn] = 0
                elif self.space_lattice[randomRow,randomColumn]==3 and self.space_lattice[neighborRow,neighborColumn]==2:
                    self.space_lattice[neighborRow,neighborColumn] = 0
            else:
                if self.space_lattice[randomRow,randomColumn]==1 and self.space_lattice[neighborRow,neighborColumn]==2:
                    self.space_lattice[randomRow,randomColumn] = 2
                    self.space_lattice[neighborRow,neighborColumn] = 1
                elif self.space_lattice[randomRow,randomColumn]==2 and self.space_lattice[neighborRow,neighborColumn]==1:
                    self.space_lattice[randomRow,randomColumn] = 1
                    self.space_lattice[neighborRow,neighborColumn] = 2
                elif self.space_lattice[randomRow,randomColumn]==1 and self.space_lattice[neighborRow,neighborColumn]==3:
                    self.space_lattice[randomRow,randomColumn] = 3
                    self.space_lattice[neighborRow,neighborColumn] = 1
                elif self.space_lattice[randomRow,randomColumn]==3 and self.space_lattice[neighborRow,neighborColumn]==1:
                    self.space_lattice[randomRow,randomColumn] = 1
                    self.space_lattice[neighborRow,neighborColumn] = 3
                elif self.space_lattice[randomRow,randomColumn]==2 and self.space_lattice[neighborRow,neighborColumn]==3:
                    self.space_lattice[randomRow,randomColumn] = 3
                    self.space_lattice[neighborRow,neighborColumn] = 2
                elif self.space_lattice[randomRow,randomColumn]==3 and self.space_lattice[neighborRow,neighborColumn]==2:
                    self.space_lattice[randomRow,randomColumn] = 2
                    self.space_lattice[neighborRow,neighborColumn] = 3
        
#         self.history.append((self.num_a, self.num_A))
#         return self.space_lattice
```

```python
# Define the width and length of lattice
N=50
# Define the frequency of initial occurrence of each: empty (0), rock (1), paper (2), scissors (3)
p_empty = 0.3
p_rock = 0.24
p_paper = 0.23
p_scissors = 0.23

space_lattice = np.array(random.choices([0,1,2,3], weights=[p_empty,p_rock,p_paper,p_scissors], k=N**2))
space_lattice = space_lattice.reshape(N,N)

p_fight = 0.99

seed = 10

box = agent_based_model(space_lattice,N,p_fight,seed)

# set up figure and animation
fig = plt.figure()
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax = fig.add_subplot(111, aspect='equal', autoscale_on=False)

cmap = colors.ListedColormap(['w','r','b','g'])

ax.imshow(box.space_lattice,cmap=cmap)


def init():
    """initialize animation"""
    global box

    return box.space_lattice

def animate(i):
    """perform animation step"""
    global box, ax, fig
    box.step(box.space_lattice)
    ax.imshow(box.space_lattice, extent=[0, 1, 0, 1], cmap=cmap)

    return box.space_lattice

ani = FuncAnimation(fig, animate, frames=500,blit=False,
                              interval=100,  init_func=init)


# save the animation as an mp4.  This requires ffmpeg or mencoder to be
# installed.  The extra_args ensure that the x264 codec is used, so that
# the video can be embedded in html5.  You may need to adjust this for
# your system: for more information, see
# http://matplotlib.sourceforge.net/api/animation_api.html
ani.save('particle_box.html')

plt.show()
```
